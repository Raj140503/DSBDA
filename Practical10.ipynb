```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score
```

```python
# 2 - Find the logistic regression for student if they study 25 and 42 hours
study_hours = np.array([29, 15, 33, 48, 39]).reshape(-1, 1)
pass_fail = np.array([0, 0, 1, 1, 1])
```

```python
model = LogisticRegression()
model.fit(study_hours, pass_fail)
```

```python
hours_to_predict = np.array([25, 42]).reshape(-1, 1)
predictions = model.predict(hours_to_predict)
```

```python
print("Prediction for 25 hours:", "Pass" if predictions[0] == 1 else "Fail")
print("Prediction for 42 hours:", "Pass" if predictions[1] == 1 else "Fail")
```

```python
data = pd.read_csv('/content/Social_Network_Ads.csv')
data
```

```python
# 3 - Compute Confusion matrix to find TP, FP, TN, FN, Accuracy, Error rate, Precision, Recall on the given dataset.
from sklearn.metrics import confusion_matrix
```

```python
y_true = pass_fail

y_pred = model.predict(study_hours)

cm = confusion_matrix(y_true, y_pred)

TN = cm[0, 0]
FP = cm[0, 1]
FN = cm[1, 0]
TP = cm[1, 1]
```

```python
accuracy = (TP + TN) / np.sum(cm)
error_rate = (FP + FN) / np.sum(cm)
precision = TP / (TP + FP)
recall = TP / (TP + FN)
```

```python
print("Confusion Matrix:")
print(cm)
print("True Positives (TP):", TP)
print("False Positives (FP):", FP)
print("True Negatives (TN):", TN)
print("False Negatives (FN):", FN)
print("Accuracy:", accuracy)
print("Error Rate:", error_rate)
print("Precision:", precision)
print("Recall:", recall)
```

```python
# 1 - Implement logistic regression using Python/R to perform classification on Social_Network_Ads.csv dataset.
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
X, y = data.iloc[:, [2, 3]], data.iloc[:, 4]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)
sc = StandardScaler().fit(X_train)
X_train, X_test = sc.transform(X_train), sc.transform(X_test)
classifier = LogisticRegression(random_state=0).fit(X_train, y_train)
y_pred = classifier.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))
```
