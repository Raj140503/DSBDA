```python
# 1 - Import all the required Python Libraries and Load the Dataset into pandas data frame.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("/content/AcademicPerformance_New.csv")
data
```

```python
# 2 - Data Preprocessing: find the missing values in the data and display statistical information.
missing_value = data.isnull().sum()
statistical_info = data.describe()

print("Missing_values: ")
print(missing_value)
print("\nStatistical Information: ")
print(statistical_info)
```

```python
# 3 - Provide variable descriptions. Types of variables etc. Check the dimensions of the data frame.
variable_description = {col: f'Description of {col}' for col in data.columns}
print("Variable Description: ")
print(variable_description)
```

```python
variable_types = data.dtypes
print("Variable Types: ")
print(variable_types)
```

```python
dimensions = data.shape
print("Dimensions of Data frames: ", dimensions)
```

```python
# 4 - Data Normalization: Perform min max normalization and plot box plot for any column.
x = data[['Average']].values.astype(float)
x
```

```python
from sklearn import preprocessing
min_max_scaler= preprocessing.MinMaxScaler()
x_scale = min_max_scaler.fit_transform(x)
data_normalized = pd.DataFrame(x_scale)
data_normalized
```

```python
data[['Average']].max()
```

```python
data[['Average']].min()
```

```python
data['Average'].plot(kind='box',title='Average of Marks')
plt.show()
```

```python
# 5 - Turn categorical variables for PG column into quantitative variables in Python.
data['PG'].replace(['Yes','No'],[0,1],inplace=True)
data
```
