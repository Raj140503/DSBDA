```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("/content/salary.csv")
data
```

```python
X = data[['experience']]
y = data['salary']
model = LinearRegression()
model.fit(X, y)
```

```python
years_of_experience = np.array([[15]])
predicted_salary = model.predict(years_of_experience)

print("Predicted salary for 15 years of experience: Rs", predicted_salary[0])
```

```python
plt.scatter(data['experience'], data['salary'], color='purple', label='Data points')
plt.plot(X, model.predict(X), color='blue', label='Regression line')
plt.xlabel('Years of Experience')
plt.ylabel('Salary (RS)')
plt.title('Salary vs. Years of Experience')
plt.legend()
plt.show()
```
