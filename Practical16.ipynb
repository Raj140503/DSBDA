```python
import pandas as pd
import numpy as np
import seaborn as sns
%matplotlib inline
import matplotlib as plt
```

```python
data=pd.read_csv('/content/Social_Network_Ads.csv')
data
```

```python
data['Gender'].replace(['Male','Female'],[0,1],inplace=True)
data
```

```python
# 1 - Implement linear regression using Python/R to perform classification on Social_Network_Ads.csv dataset.
x=data[["User ID","Gender","Age","EstimatedSalary"]]
y=data["Purchased"]
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size =0.2, random_state = 0)
```

```python
print("xtrain shape : ", xtrain.shape)
print("xtest shape  : ", xtest.shape)
print("ytrain shape : ", ytrain.shape)
print("ytest shape  : ", ytest.shape)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(xtrain,ytrain)
```

```python
y_pred = model.predict(xtest)
y_pred
```

```python
# 2 - Find the linear regression for student if he score 78 and 56 in 10th find score they get in 12th exam
from sklearn.linear_model import LinearRegression

X = [[90], [80], [75], [70], [60]]
y = [80, 90, 75, 65, 70]

model = LinearRegression()
model.fit(X, y)

predicted_12th_score_78 = model.predict([[78]])
predicted_12th_score_56 = model.predict([[56]])

print("Predicted 12th grade score for 10th grade score of 78:", predicted_12th_score_78[0])
print("Predicted 12th grade score for 10th grade score of 56:", predicted_12th_score_56[0])
```
